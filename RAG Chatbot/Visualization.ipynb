```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_grid(ax, x, y):
    for i in range(len(y)):
        for j in range(len(x)):
            rect = patches.Rectangle((j, i), 1, 1, linewidth=1, edgecolor='black', facecolor='lightblue')
            ax.add_patch(rect)

fig, ax = plt.subplots(figsize=(14, 10))

# Draw the grid
draw_grid(ax, range(6), range(6))

# Add text tokens within the grid cells
tokens = ["'How'", "'does'", "'Betizen'", "'handle'", "'returns'", "'?'"]
for i, token in enumerate(tokens):
    ax.text(0.5, 5.5 - i, token, fontsize=12, va='center', ha='center')

# Step 2: Compute Hidden States
# Adding hidden states text
hidden_states = ["Hidden State 1", "Hidden State 2", "Hidden State 3", "Hidden State 4", "Hidden State 5"]
for i, state in enumerate(hidden_states):
    ax.text(1 + i, 6.7, state, fontsize=10, va='center', ha='center')

# Step 3: Combine Multiple Outputs from Multiple Attention Heads
# Adding attention heads
attention_heads = ["Attention Head 1", "Attention Head 2", "Attention Head 3"]
for i, head in enumerate(attention_heads):
    ax.text(6.5, 5.5 - i, head, fontsize=10, va='center', ha='center')
    rect = patches.Rectangle((6, 5 - i), 1, 1, linewidth=1, edgecolor='black', facecolor='yellow')
    ax.add_patch(rect)

# Adjust limits and turn off the axis
ax.set_xlim(0, 8)
ax.set_ylim(0, 7.5)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Create data for heatmap
data = np.random.rand(6, 6)

fig, ax = plt.subplots(figsize=(10, 8))

# Create a heatmap
sns.heatmap(data, annot=True, cmap='YlGnBu', cbar=True, xticklabels=["'How'", "'does'", "'Betizen'", "'handle'", "'returns'", "'?'"], 
            yticklabels=["'How'", "'does'", "'Betizen'", "'handle'", "'returns'", "'?'"], linewidths=.5, ax=ax)

# Titles and labels
plt.title('Attention Scores Heatmap')
plt.xlabel('Tokens')
plt.ylabel('Tokens')

plt.show()
```
